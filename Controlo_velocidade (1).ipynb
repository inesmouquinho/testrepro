## <span style = "color:orange">Controlo de velocidade de um carro</span>

Considere-se a equação  da física clássica $f(t)=m.a(t)$, a qual descreve para cada instante $t$, a dinâmica ou aceleração $a(t)$ de um corpo de massa $m$ sujeito a uma força aplicada $f(t)$.

Com esta dinâmica, pretende-se controlar a velocidade $\color{orange}{v(t)}$ de um carrinho com massa $\color{orange}{m}$ sujeito a uma força $\color{orange}{f(t)}$ proporcioal ao erro de velocidade relativo à velocidade de referência $\color{orange}{Ref}$ pretendida e a uma constante de proporcionalidade de $\color{orange}{K_p}$  tal que: $\color{orange}{f(t) = K_p(Ref - v(t))}$

O modelo dinâmico (com equações diferênciais) deste sistema é dado por:

$\hspace{8mm}\boxed{K_p.Ref - K_p.v(t) = m.v'(t)}\hspace{4mm}(1)$

considere os seguintes parametros do problema que:
>- $K_p = 1$, constante de proporcionalidade;
>- $m = 1$, massa do carro.
>- $h = 0.1$, periodo de amostragem
>- $ref = 1$, aplique o escalão unitário como sinal de referência.
>- $v_0 = 0$, considere que a velocidade inicial é "zero"

---

## Construa o gráfico dos sinais $erro(t)$, $f(t)$ e $v(t)$ no intervalo $[0;2]$

```python
# Parâmetros do problema
m = 1
Kp = 2
h = 0.1
ref = 1
v0 = 0

tmin = 0
tmax = 2
npmin = int(tmin/h)
npmax = int(tmax/h)

t = [i*h for i in range(npmin,npmax+1)]
#print(t)
```

```python
# Funções do problema

def erro(i):
    e.append(ref - v[i])
    return e[i]

def forca(i):
    f.append(Kp*erro(i))
    return f[i]

def velocidade(i):
    return forca(i)*h/m+v[i]

e=[]
f=[]
v=[v0]
for i in range(npmin,npmax+1):
    v.append(velocidade(i))
    
# Gráfico da função
import matplotlib.pyplot as plt
plt.plot(t[:],e[:],'r--')
plt.plot(t[:],f[:],'g--')
plt.plot(t[:],v[:-1],'b--')
plt.xlabel('tempo [s]')
plt.ylabel('e, f, v')
plt.title('Controlo de velocidade')
# Add a Legend
plt.legend(['erro','força','velocidade'])
plt.grid(True)
plt.show()
```
